```python
import pandas as pd
```

```python
html = pd.read_html('https://en.wikipedia.org/wiki/List_of_Super_Bowl_champions')
```

```python
championships = html[1][0:53]
```

```python
championships.head()
```

```python
date_format = championships['Date/Season'].apply(lambda x: x[0:16])
championships.date = pd.to_datetime(date_format)
# date_column = pd.to_datetime(date_format)
```
